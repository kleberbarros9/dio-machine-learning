```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# gerando os dados
X, y = make_regression(n_samples=200, n_features=1, noise=30)

# criando o modelo de regress達o linear
model = LinearRegression()

# treinando o modelo
model.fit(X, y)

# vizualizando os dados e linha da regress達o
plt.scatter(X, y, label='Dados originais')
plt.plot(X, model.predict(X), color='red', label='Regress達o linear')
plt.title('Regress達o Linear com Scikit-Learn')
plt.xlabel('Variavel X (preditor)')
plt.ylabel('Variavel y (targer)')
plt.legend()
plt.show()
```
