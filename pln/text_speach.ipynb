```python
# Importa a classe gTTS (Google Text-to-Speech) da biblioteca gTTS
from gtts import gTTS

# Define o texto que será convertido em fala
text_to_say = "How are you doing?"

# Define o idioma da fala, onde "en" é para inglês
language = "en"

# Cria um objeto gTTS, que irá gerar a fala a partir do texto fornecido
# O parâmetro 'slow' define se a fala será lenta ou rápida (False para normal)
gtts_object = gTTS(text = text_to_say,
                   lang= language,
                   slow= False)

# Salva o áudio gerado em um arquivo WAV no diretório especificado
gtts_object.save("/content/gtts.wav")
```

```python
from IPython.display import Audio

Audio("/content/gtts.wav", autoplay=True)
```

```python
# Define o texto em francês que será convertido em fala
french_text = "Je vais au supermarché aujourd'hui"

# Define o idioma da fala, onde "fr" é para francês
french_language = "fr"

# Cria um objeto gTTS para gerar a fala a partir do texto em francês
# O parâmetro 'slow' é definido como True, o que fará a fala ser mais lenta
french_gtts_object = gTTS(text = french_text,
                          lang= french_language,
                          slow= True)

# Salva o áudio gerado em um arquivo WAV no diretório especificado
french_gtts_object.save("/content/french.wav")

# Reproduz o áudio gerado
Audio("/content/french.wav")
```
