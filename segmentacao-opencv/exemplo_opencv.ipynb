```python
import numpy as np
import cv2 as cv
import os

# Importa a função necessária para exibir imagens no Colab
from google.colab.patches import cv2_imshow

# Faz o download do arquivo haarcascade se ele não existir
if not os.path.exists('haarcascade_frontalface_default.xml'):
    !wget https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml

# Carrega o classificador de faces
face_classifier = cv.CascadeClassifier('haarcascade_frontalface_default.xml')

# Carrega a imagem
image = cv.imread('image3.jpg')

# Verifica se a imagem foi carregada com sucesso
if image is None:
    raise ValueError("Não foi possível ler a imagem. Verifique o caminho do arquivo.")

# Converte a imagem para tons de cinza para a detecção de faces
image_gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

# Detecta faces na imagem em tons de cinza
faces = face_classifier.detectMultiScale(image_gray, 1.3, 5)

# Desenha retângulos em torno das faces detectadas
for (x, y, w, h) in faces:
    cv.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)

# Exibe a imagem resultante com as faces detectadas
cv2_imshow(image)
```
