```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Importa a imagem "tiger.jpg" em formato colorido (BGR).
img = cv2.imread('tiger.jpg')

# Converte a imagem carregada para escala de cinza (grayscale).
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Aplica um filtro de desfoque Gaussiano na imagem em escala de cinza,
# utilizando um kernel de 3x3 para suavizar ruídos e reduzir detalhes.
img = cv2.GaussianBlur(gray, (3,3), 0)
```

```python
# Aplica o operador Sobel na direção x (horizontal).
# - src=img: utiliza a imagem suavizada como entrada.
# - ddepth=cv2.CV_64F: define a profundidade do resultado como 64 bits de ponto flutuante.
# - dx=1, dy=0: deriva a imagem em relação a x (horizontal).
# - ksize=5: especifica o tamanho do kernel como 5x5.
sobelx = cv2.Sobel(src=img, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5)

# Aplica o operador Sobel na direção y (vertical).
# - dx=0, dy=1: deriva a imagem em relação a y (vertical).
sobely = cv2.Sobel(src=img, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5)

# Aplica o operador Sobel para derivar a imagem nas direções x e y simultaneamente.
# - dx=1, dy=1: deriva a imagem em ambas as direções.
sobelxy = cv2.Sobel(src=img, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5)
```

```python
# Configura o tamanho da figura para visualização usando Matplotlib.
plt.figure(figsize=(18, 19))

# Subplot 1: Exibe a imagem original (em escala de cinza) no canto superior esquerdo.
# - cmap='gray': define o mapa de cores como escala de cinza.
# - title: adiciona o título "Original".
# - axis('off'): remove os eixos para uma apresentação mais limpa.
plt.subplot(221)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')

# Subplot 2: Exibe o resultado do operador Sobel nas direções X e Y no canto superior direito.
# - Título: "Sobel X Y using Sobel()".
plt.subplot(222)
plt.imshow(sobelxy, cmap='gray')
plt.title('Sobel X Y using Sobel()')
plt.axis('off')

# Subplot 3: Exibe o resultado do operador Sobel na direção X no canto inferior esquerdo.
# - Título: "Sobel X".
plt.subplot(223)
plt.imshow(sobelx, cmap='gray')
plt.title('Sobel X')
plt.axis('off')

# Subplot 4: Exibe o resultado do operador Sobel na direção Y no canto inferior direito.
# - Título: "Sobel Y".
plt.subplot(224)
plt.imshow(sobely, cmap='gray')
plt.title('Sobel Y')
plt.axis('off')

# Exibe todas as subplots configuradas.
plt.show()
```
